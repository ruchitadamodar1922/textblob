```python
import pandas as pd
data=pd.read_csv('shop1_review.txt',sep='\t',header=None)
nameColumn=['review','sentiment analysis']
data.columns=nameColumn
data.head()
```

```python
from textblob import TextBlob
pol1=lambda x: TextBlob(x).sentiment.polarity
sub1=lambda x: TextBlob(x).sentiment.subjectivity

data['polarity']=data['review'].apply(pol1)
data['subjectivity']=data['review'].apply(sub1)
data
```
